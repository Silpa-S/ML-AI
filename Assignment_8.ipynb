Importing Libraries

```python

import pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
```

Reading Dataset

```python
df=pd.read_csv('/content/judge-1377884607_tweet_product_company.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

Dropping column

```python
df.drop('emotion_in_tweet_is_directed_at',axis=1,inplace=True)
```

Preprocessing

```python
# Convert 'tweet_text' column to strings
df['tweet_text'] = df['tweet_text'].astype(str)

# Preprocessing
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['tweet_text'])
sequences = tokenizer.texts_to_sequences(df['tweet_text'])
X = pad_sequences(sequences)
y = pd.get_dummies(df['is_there_an_emotion_directed_at_a_brand_or_product']).values
```

Model

```python
# Model Building
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=X.shape[1]))
model.add(LSTM(units=64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(units=4, activation='softmax'))
```

```python
model.summary()
```

Compiling the model

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

Training the model

```python
# Training
model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)
```
